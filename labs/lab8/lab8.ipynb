```python
import numpy as np
import matplotlib.pyplot as plt
import math
from numpy.linalg import inv, norm
```

# Question 1 Linear Spline


**Subroutines implmented in eval_lin_spline.py**

```python
from eval_lin_spline import LinearSplineEvaluator
```

```python
def f(x):
    return 1 / (1 + (10 * x) ** 2)


N = 10 
h = 2 / (N - 1)

x_nodes = np.array([-1 + (j - 1) * h for j in range(N)])
y_nodes = f(x_nodes)

x_eval = np.linspace(-1, 1, 200)

evaluator = LinearSplineEvaluator(
    interval_start=-1, interval_end=1, function=f, num_intervals=N-1)
y_eval = evaluator.evaluate(x_eval)


error = abs(y_eval-f(x_eval))
```

```python

plt.figure()
plt.plot(x_eval, f(x_eval), label="True Function",
        linestyle="dashed", color="black")
plt.plot(x_eval, y_eval, label="Linear Spline Approximation", color="red")
plt.scatter(x_nodes, y_nodes, color="blue",
            label="Interpolation Nodes")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Linear Spline Interpolation of f(x)")
plt.legend()
plt.grid()
plt.show()


plt.figure()
plt.plot(x_eval,error,'r--', label= "Error Graph" )
plt.grid()
plt.legend()
plt.title('Error of linear spline')
plt.show()
    

```

# Question 2 Cubic Spline

**Subroutines implemented in cubic_spline.py**

```python
from cubic_spline import CubicSplineInterpolator
```

```python
interpolator = CubicSplineInterpolator(f, -1, 1, N)

x_eval_cubic= np.linspace(-1, 1, 100)

y_eval_cubic = interpolator.evaluate(x_eval_cubic)

f_x_eval_cubic = np.array(f(x_eval_cubic))

error_cubic = abs(y_eval_cubic - f_x_eval_cubic)
```

```python
plt.figure()
plt.plot(x_eval_cubic, f_x_eval_cubic,
         label='f(x)', color="black")

plt.plot(x_eval_cubic, y_eval_cubic, 
         'b--', label='Cubic Spline')


plt.legend()
plt.title('Function and Spline Approximation')
plt.grid()
plt.show()



plt.plot(x_eval_cubic, error_cubic, 'r--')
plt.title('Error of Cubic Spline')
plt.grid()
plt.show()
```
